```python
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv("C:\\Users\\Stemadmin\\Desktop\\Anuj Khadka\\carbon_footprint\\results\\benchmark_20260424_003610.csv")

# filter to large size only
large = df[df['size'] == 'large']

# pull the two groups
c_data     = large[large['language'] == 'c']['gco2e'].values
python_data = large[large['language'] == 'python']['gco2e'].values

# run the test
t, p = stats.ttest_ind(c_data, python_data, equal_var=False)

print(f"t = {t:.3f}")
print(f"p = {p:.6f}")

if p < 0.0033:
    print("Significant!")
else:
    print("Not significant.")
```

```python
from itertools import combinations
import numpy as np

langs = ['c', 'rust', 'go', 'java', 'javascript', 'python']
alpha = 0.05 / 15  # Bonferroni correction

for l1, l2 in combinations(langs, 2):
    a = large[large['language'] == l1]['gco2e'].values
    b = large[large['language'] == l2]['gco2e'].values
    
    t, p = stats.ttest_ind(a, b, equal_var=False)
    
    result = "SIGNIFICANT" if p < alpha else "not significant"
    print(f"{l1} vs {l2}: t={t:.3f}, p={p:.4f} → {result}")
```
